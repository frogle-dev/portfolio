### Changes to UFO movement:

First of all, we modified the movement of the UFO. We thought that the previous movement was sort of repetetive. So, we decided to add horizontal movement on top of the vertical oscillation. 
Our first iteration consisted of making the UFO move horizontally with the same function as vertically. This created a sort of diagonal oscillation effect.
Next, we tried changing the strength and speeds of each movement to vary it more.
Finally, we settled on changing the function of the horizontal movement to a tangent function. This created a cool effect of the UFO teleporting from one side of the screen to the other.

Here was our code for this:
```
class Player extends GameObject {
	constructor(image, gameWorld) {
	const width = image.naturalWidth / 2;
	const height = image.naturalHeight / 2;
	const x = (gameWorld.width - width) / 2;
	const y = (gameWorld.height - height) / 2;
	super(image, width, height, x, y);
	this.baseY = y;
	this.baseX = x;
	this.frame = 0;
	}
	update() {
	canvas.onkeydown = function(evt) {
		changeDir(evt.keyCode);
	}

	this.x = this.baseX + Math.tan(this.frame * 0.1) * 40;
	this.y = this.baseY + Math.sin(this.frame * 0.05) * 20;
	this.frame++;
	}
}
```

The line we specifically changed was this:
```
this.x = this.baseX + Math.tan(this.frame * 0.1) * 40;
```
Essentially, this takes the UFO's x position and offsets it by the tangent function. This function takes in the current frame as input in order to make it move as the time progresses. Then, the "* 0.1" is used to slow down its movement and the "* 40" is used to expand the distance it covers.
