<h2 style="color: black !important;">Terminal</h2>

<div>
<b>What is the terminal?:</b><br/><br/> Put simply, the terminal is a direct way of interacting with the computer through text input and outputs.

The terminal can be confusing at first for many beginners in this class, especially with all the different commands needed for different functions. It can be hard to remember what the commands are, and what they do. So, this interactive study guide aims to help you understand each terminal command so you wont forget them.<br/><br/>This guide will provide you with both information and practice challenges to make sure you've got everything on lock.<br/><br/>

We will talk about:
<ul>
    <li>directories (and what they are)</li>
    <li>directory management</li>

</ul>

<h3 style="color: black !important;">What are directories?</h3>

<b>What?:</b> Directories are an organized way for the computer (and the user) to locate and use files throughout the system. This is typically structured as a tree of folders which each contain different types of files.

<b>Why?:</b> Directories are an efficient and intuitive way to navigate around the computer which is why it can be used as an interface for both the user and the computer itself.

Even though directories might seem complicated at first, after practicing and understanding it, they actually become really easy to work with.

Think about directories as a tree: <br/>
<img src="{{site.baseurl}}/images/tools/sci-fi-tool/file-tree.png" alt="Directory Tree"><br/>
Each node in the tree represents a directory (folder), this directory then contains a few files of its own, but also contains other folders. Those folders then do the same thing. So, this ends up creating a large complex of folders and files, but also stays organized and navigable.

<h3 style="color: black !important;">How to manage your directories</h3>

Now that you know the structure of directories, you need to know how to actually navigate through them. <br/>
Most people are already familiar with the visual file explorers that most OSes provide, but if you want to start interacting deeper with directories, you have to know how to navigate them in the terminal.

The typical way to signify the location of a file or folder is to use paths, which is a string of the names of each folder/file separated by slashes <br/>
For example: <br/>
<code>User/MyStuff/Music/Jazz/jazz_song.mp3</code><br/><br/>
This path goes from the top folder "User" -> "MyStuff" -> "Music" -> "Jazz" -> "jazz_song.mp3" <br/>
So, "User", "MyStuff", "Music", "Jazz" are all folders because they contain other folders and files, and "jazz_song.mp3" is a file because it only stores its own data and ends with the file extension ".mp3" <br/><br/>
<i>Also note that paths are relative from the current folder you are in.</i> <br/><br/>

<b>Commands</b><br/>
There are a few simple commands you need to know in order to move around in terminal space: <br/>
<code>
cd
</code>
<ul>
	<li>Moves you to a different directory</li>
    <li>Probably most commonly used command</li>
    <li>Stands for "change directory"</li>
    <li>Syntax: "cd {path to your directory}"</li>
	<li>You can also write "cd .." to move back a folder</li>
</ul>
<br/>
<code>
pwd
</code>
<ul>
	<li>States current path to your directory</li>
		<ul>
			<li>ex (in terminal): <code>User/Mystuff/Mymusic: pwd<br/>Home/User/Mystuff/Mymusic</code></li>
		</ul>
	<li>Prints full name of current directory</li>
</ul>
<br/>
<code>
./
</code>
<ul>
	<li>Runs batch files</li>
	<li>For linux, it uses .sh</li>
	<li>Ex: <code>./scripts/venv.sh</code></li>
	<ul>
		<li>Venv is the batch file</li>
		<li>Scripts is the folder venv is stored it</li>
	</ul>
</ul>
<br/>
<code>
rm
</code>
<ul>
	<li>Deletes a file</li>
	<li>Stands for "remove"</li>
	<li>Syntax: "rm {path file/folder you want to remove}</li>
	<li>"rm -r" to delete a folder (-r means recursive)</li>
	<li>"rm -f" to delete a file without warnings (can be risky)</li>
	<li>"rm -rf" to delete a folder without warnings (can be risky)</li>
</ul>
<br/>
<code>
mv
</code>
<ul>
	<li>Moves directories to different folders or directories</li>
	<ul>
		<li>Ex: <code>mv thisfolder otherfolder</code></li>
		<li><code>thisfolder</code> is the folder you are trying to move; <code>otherfolder</code> is the folder you are moving it into</li>
	</ul>
	<li>Renames folders/directories</li>
	<ul>
		<li>Ex: <code>mv thisfolder newfolder</code></li>
		<li><code>thisfolder</code> is the original name of the folder; <code>newfolder</code> is the new name of the folder (the new name cannot be an already existing folder)</li>
	</ul>
</ul>
<br/>
<code>
mkdir
</code>
<ul>
	<li><code>mkdir</code> is a command standing for "make directory"</li>
	<li>Makes a new directory that you can put other directories/folders in</li>
</ul>

<h3 style="color: black !important;">Git Clone</h3>

<b>What is Git Clone?</b><br/><br/>

Git Clone is a command that can be used in terminal that clones an existing repository for github. It creates the repository you cloned as a new directory. It downloads the entire repository, meaning it also includes the history of commits, branches, and files.<br/>
Ex: <code><span>git clone [https]://github.com/username/repository-name.git</span></code><br/>
If the repository you cloned is something your github account has access to, you can:
 - <b>Pull</b> updates on the remote repository to sync changes<br/>
 - <b>Push</b> your changes back to the Github Repo<br/>
